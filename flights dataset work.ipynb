### Q2. Write a function that given your birthday (as a date), returns how old you are in years.

```python
import datetime
currentdate = datetime.date.today()
dob = input('Date of Birth : ')
birth = datetime.datetime.strptime(dob, "%Y-%m-%d").date()
diff = currentdate.year - birth.year
print(diff,' years old')
```
